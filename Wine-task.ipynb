```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_wine
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split, KFold
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
info = load_wine()
info
```

```python
print(info.DESCR)
```

```python
x = pd.DataFrame(info.data, columns=info.feature_names)
x
```

```python
y = info.target
y
```

```python
x.info()
```

```python
x[["alcohol", "malic_acid", "ash", "alcalinity_of_ash", "magnesium", "total_phenols", "flavanoids", "nonflavanoid_phenols", "proanthocyanins", "color_intensity", "hue", "od280/od315_of_diluted_wines", "proline"]] = x[["alcohol", "malic_acid", "ash", "alcalinity_of_ash", "magnesium", "total_phenols", "flavanoids", "nonflavanoid_phenols", "proanthocyanins", "color_intensity", "hue", "od280/od315_of_diluted_wines", "proline"]].astype("float16")
```

```python
x.info()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25,  random_state=42)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
scaler = MinMaxScaler()

x_train_sc = pd.DataFrame(scaler.fit_transform(x_train), columns=x_train.columns, index=x_train.index)
x_test_sc = pd.DataFrame(scaler.transform(x_test), columns=x_test.columns, index=x_test.index)
```

```python
np.unique(y, return_counts=True)
```

```python
model = LogisticRegression(solver="newton-cg")
model.fit(x_train, y_train)
```

```python
pred_train = model.predict(x_train)
```

```python
pred_test = model.predict(x_test)
```

```python
accuracy_score(y_train, pred_train), accuracy_score(y_test, pred_test)
```
